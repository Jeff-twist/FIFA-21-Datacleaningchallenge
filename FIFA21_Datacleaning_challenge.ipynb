##### Next step is to write functions to convert these irregular values to a standard and consistent one for each columns.

##### Now let's write a function to clean up the 'Height' column

```python
# creating a function to convert height to metres
def metre_conversion(Height):
    if 'cm' in Height:
        return float(Height.replace('cm', '')) / 100
    else:
        feet, inches = Height.split('\'')
        return round(float(feet) * 0.3048, 2) + round(float(inches.replace('"', '')) * 0.0254, 2)
    
    
# applying the function to the height column
fifa['Height(m)'] = fifa['Height(m)'].apply(metre_conversion)

# Checking for change 
fifa['Height(m)'].unique()
```

##### Now let's write a function to clean up the 'Weight' column

```python
# create a function to convert weight to kilogram
def kg_conversion(weight):
    if 'kg' in weight:
        return (weight.replace('kg', '')) 
    else:
        return round(float(weight.replace('lbs', '')) /2.20462, 2)
    
    
#applying the function to the weight column
fifa['Weight(kg)'] = fifa['Weight(kg)'].apply(kg_conversion)

# Checking for change  
fifa['Weight(kg)'].unique()
```

##### Now let's write a function to clean up the 'Wage', 'Release clause', and 'Value' columns

```python
# define a function to remove '€' and convert the 'K' and 'M' to thousands and millions respectively
def convert_value(value_str):
    value = value_str.replace('€', '')
    if value_str.endswith ('K'):
        value = float(value[:-1]) * 1000
    elif value_str.endswith ('M'):
        value = float(value[:-1]) * 1000000
    else:
        value = float(value)
    if value.is_integer():
        return '{:.0f}'.format(value)
    else:
        return '{:.2f}'.format(value)
```

```python
# apply the function to the wage, value and release clause columns
fifa[['Value', 'Wage', 'Release Clause']] = fifa[['Value', 'Wage', 'Release Clause']].applymap(convert_value)
```

```python
fifa[['Value', 'Wage', 'Release Clause']]
```

##### Next is to clean up the Club column that has '\n\n'

```python
#stripping the unnecessary characters in the club column
fifa['Club'] = fifa['Club'].str.replace('\n\n\n\n','')
fifa.head()
```

##### Next is to clean up the star symbols that appear in the 'Weak_foot_rating', 'Skill_moves_rating', and 'Injury_rating' columns

```python
#stripping the star character in columns Weak_foot_rating, Skill_moves_rating and Injury_rating
fifa[['Weak_foot_rating','Skill_moves_rating','Injury_rating']] = fifa[['Weak_foot_rating','Skill_moves_rating','Injury_rating']].apply(lambda x: x.str.replace('★',''))
fifa.head()
```

##### Next is to resolve the contract column by creating a new column called 'Contract_status' which shows players on active contract, on loan and free agents. Two new columns would be created for contract start year and contract end year for start and end of contract respectively.

```python
# splitting the contract column and creating 2 separate Contract_start_year and Contract_end_year columns
fifa[['Contract_start_year', 'Contract_end_year']] = fifa['Contract'].str.split('~', expand=True)

fifa.head()
```

##### Next is to categorize players based on active contract, loan or free agents

```python
# Creating the categories
conditions = [
    fifa['Contract'] == 'Free',
     fifa['Contract'].str.contains('On Loan')
]
values = ['Free agent', 'On Loan']

# using the numpy '.select()' to create a new column based on conditions and values
fifa['Contract_status'] = np.select(conditions, values, default = 'Active')

# display the result
fifa.head()
```

```python
# Using the year in the contract column as contract_end_year for players on loan
fifa.loc[fifa['Contract_status'] == 'On Loan', 'Contract_end_year'] = fifa['Contract'].apply(lambda x: x.split(',')[-1].strip()[:4])
```

```python
# dropping the contract and contract_start_year columns
fifa.drop(columns = ['Contract', 'Contract_start_year'], inplace = True)
fifa.head()
```

##### Next is to handle the null/missing values and figures that contains 'K' in the hits column.


```python
# creating a function to convert hits column
def convert_hits(hit):
    if pd.isna(hit):
        return 0
    elif isinstance(hit, int):
        return hit
    elif 'K' in hit:
        return int(float(hit.replace('K', '')) * 1000)
    else:
        return int(float(hit))

# applying the function the Hits column
fifa['Hits'] = fifa['Hits'].apply(convert_hits)
```

##### Next is to convert some column data types to the appropriate data types

```python
# converting Weight and  Wage to integer
fifa[['Weight(kg)', 'Wage']] = fifa[['Weight(kg)', 'Wage']].astype(int) 

# converting Value and Release_clause to float
fifa[['Release Clause', 'Value']] = fifa[['Release Clause', 'Value']].astype(float)

# Converting ID column to string
fifa['ID'] = fifa['ID'].astype(str)

# converting Joined and Contract_end_year to datetime
fifa['Joined'] = pd.to_datetime(fifa['Joined'])

# converting contract_status to category
fifa[['Contract_status', 'Skill_moves_rating']] = fifa[['Contract_status', 'Skill_moves_rating']].astype('category')
```

```python
# Converting Weak_foot_rating, Skill_moves_rating, Injury_rating, Attacking_work_rate and Defensive_work_rate to ordered category
ordinal_dict = {'Attacking_work_rate': ['Low', 'Medium', 'High'], 'Defensive_work_rate': ['Low', 'Medium', 'High'],
               'Weak_foot_rating': ['1 ', '2 ', '3 ', '4 ', '5 '], 'Injury_rating': ['1 ', '2 ', '3 ', '4 ', '5 ']}
for key, value in ordinal_dict.items():
    order = pd.api.types.CategoricalDtype(categories = value, ordered = True)
    fifa[key] = fifa[key].astype(order)
```

##### Based on the information on the FIFA website, attributes such as Attacking, Defending, Skill, Power, Defending, Mentality, Movement and Goalkeeping are the sum of other stats under them. Therefore, suming these up and dividing by 100 gives it a more consistent value.
* Attacking = Crossing + Heading Accuracy + Finishing + Short Passing + Short Passing + Volleys
*Skill = Dribbling + FK Accuracy + Curve + Long Passing + Ball Control
*Power = Shot Power + Jumping + Stamina + Strength + Long Shots
*Defending = Standing Tackle + Sliding Tackle + Marking
*Mentality = Aggression + Interceptions + Positioning + Vision + Penalties + Composure
*Movement = Acceleration + Sprint Speed + Agility + Reactions + Balance
*Goalkeeping = GK Diving + GK Handling + GK Kicking + GK Positioning + GK Reflexes
*Total Stats = Attacking + Skill + Power + Defending + Mentality + Movement + Goalkeeping




```python
fifa.columns
```

```python
# summing up each stats
fifa["Attacking"] = fifa2["Crossing"] + fifa2["Heading Accuracy"] + fifa2["Finishing"] + fifa2["Short Passing"] + fifa2["Volleys"]
fifa["Skill"] = fifa2["Dribbling"] + fifa2["FK Accuracy"] + fifa2["Curve"] + fifa2["Long Passing"] + fifa2["Ball Control"]
fifa["Power"] = fifa2["Shot Power"] + fifa2["Jumping"] + fifa2["Stamina"] + fifa2["Strength"] + fifa2["Long Shots"]
fifa["Defending"] = fifa2["Standing Tackle"] + fifa2["Sliding Tackle"] + fifa2["Marking"]
fifa["Mentality"] = fifa2["Aggression"] + fifa2["Interceptions"] + fifa2["Positioning"] + fifa2["Vision"] + fifa2["Penalties"] + fifa2["Composure"]
fifa["Goalkeeping"] = fifa2["GK Diving"] + fifa2["GK Handling"] + fifa2["GK Kicking"] + fifa2["GK Positioning"] + fifa2["GK Reflexes"]
fifa["Movement"] = fifa2["Acceleration"] + fifa2["Sprint Speed"] + fifa2["Agility"] + fifa2["Reactions"] + fifa2["Balance"]
```

```python
# rounding each of the main skills / 100
fifa["Attacking"] = round(fifa["Attacking"]/5)
fifa["Skill"] = round(fifa["Skill"]/5)
fifa["Movement"] = round(fifa["Movement"]/5)
fifa["Power"] = round(fifa["Power"]/5)
fifa["Mentality"] = round(fifa["Mentality"]/6)
fifa["Defending"] = round(fifa["Defending"]/3)
fifa["Goalkeeping"] = round(fifa["Goalkeeping"]/5)
```

```python
# Checking to view changes
fifa.head()
```

##### Lastly, we Create a new column known as Body_mass_index (BMI) which can be used to evaluate a footballer's health and fitness. Formular for BMI is weight(kg) / (height(m))^2

```python
# create a BmI column using the formular
fifa['Body_mass_index'] = fifa['Weight(kg)'] / (fifa['Height(m)'] ** 2)

# round the BMI column to two decimal places
fifa['Body_mass_index'] = fifa['Body_mass_index'].round(2)
```
